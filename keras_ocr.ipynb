```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt 
import pytesseract
```

```python
import os
from pdf2image import convert_from_path

def convert_pdf_jpg(path_dir):
    for pdf_file in os.listdir(path_dir):

        if pdf_file.endswith(".pdf"):

            pages = convert_from_path(pdf_file, 300)
            pdf_file = pdf_file[:-4]

            for page in pages:

               page.save("%s-page%d.jpg" % (pdf_file,pages.index(page)), "JPEG")

# path_pdf=r'c:\\Users\\Admin\\Desktop\\ComputerVision\\OCR\\dataapp'
# convert_pdf_jpg(path_pdf)
```

```python
def preprocessing_2(image):
    img=cv.imread(image)
    #img_gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
    #image_tresh=cv.threshold(img_gray,0,255,cv.THRESH_BINARY_INV + cv.THRESH_OTSU)[1]
    return img

def preprocessing_3(image):
    img = cv.imread(image)
    #cv.imshow("Original image",img)
    # CLAHE (Contrast Limited Adaptive Histogram Equalization)
    clahe = cv.createCLAHE(clipLimit=3., tileGridSize=(8,8))

    lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)  # convert from BGR to LAB color space
    l, a, b = cv.split(lab)  # split on 3 different channels

    l2 = clahe.apply(l)  # apply CLAHE to the L-channel

    lab = cv.merge((l2,a,b))  # merge channels
    img2 = cv.cvtColor(lab, cv.COLOR_LAB2BGR)  # convert from LAB to BGR (image constraste)
    return img2


def display(im_path):
    dpi = 50
    im_data = plt.imread(im_path)

    height, width  = im_data.shape[:2]
    
    # What size does the figure need to be in inches to fit the image?
    figsize = width / float(dpi), height / float(dpi)

    # Create a figure of the right size with one axes that takes up the full figure
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0, 0, 1, 1])

    # Hide spines, ticks, etc.
    ax.axis('off')

    # Display the image.
    ax.imshow(im_data, cmap='gray')

    plt.show()

def display_img(title, img):
    """Displays an image on screen and maintains the output until the user presses a key"""
    cv.namedWindow('img', cv.WINDOW_NORMAL)
    cv.setWindowTitle('img', title)
    cv.resizeWindow('img', 1200, 900)
    # Display Image on screen
    cv.imshow('img', img)
    # Mantain output until user presses a key
    cv.waitKey(0)
    # Destroy windows when user presses a key
    cv.destroyAllWindows()
```

```python
import keras_ocr
import pandas as pd 

path='dataapp/test_page_0.jpg'
img1=preprocessing_2(path)
img2=preprocessing_3(path)

def x_keras_ocr(image):
    pipeline = keras_ocr.pipeline.Pipeline()
    results = pipeline.recognize([image])
    data_process_keras=pd.DataFrame(results[0], columns=['text','bbox'])

    return results,data_process_keras

```

```python
results_process_keras_1,data_process_keras_1=x_keras_ocr(img2)
data_process_keras_1.to_excel('dataapp/preprocess_3/keras_ocr_process_1.xlsx')
data_process_keras_1
```

```python
ocr_img_1 = results_process_keras_1[0]
for text, box in ocr_img_1:
    print(text)
```
