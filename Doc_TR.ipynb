Detection 

```python
from doctr.models import detection_predictor
predictor_1 = detection_predictor('db_resnet50')
document = DocumentFile.from_images("drive/MyDrive/test_page_0.jpg")
result_predictor_1=predictor_1(document)
result_predictor_1
```

```python
from doctr.models import detection_predictor
predictor_2 = detection_predictor('db_mobilenet_v3_large')
document = DocumentFile.from_images("drive/MyDrive/test_page_0.jpg")
result_predictor_2=predictor_2(document)
result_predictor_2
```

```python
#resnet + crnnVgg16
model_1 = ocr_predictor('db_resnet50', 'crnn_vgg16_bn', pretrained=True)
document_1 = DocumentFile.from_images("drive/MyDrive/crop.jpg")
result_1 = model_1(document_1)
result_1.show(document_1)
```

```python
json_response_1 = result_1.export()
json_response_1
```
